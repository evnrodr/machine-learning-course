## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4, 5, 3, 1]
y = [10, 15, 20, 15, 10, 15, 10]

# 2. Setup plot
fig, ax = plt. subplots(figsize=(5, 5)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title= "Simple Plot",
       xlabel= "x-axis",
       ylabel= "y-axis")

# Save & show (you save the whole figure)
fig.savefig('../images/sample-plot.png')
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a dictionary 
covid_deaths = {'So Paulo': 3709,
                'Rio de Janeiro': 1714,
                'Cear': 1114,
                'Pernambuco': 1047,
                'Amazonas': 1004}

fig, ax = plt.subplots(figsize=(8, 8))
ax.bar(covid_deaths.keys(), covid_deaths.values());
ax.set(title = "COVID-19 Brazil: States with the highest number of deaths",
       xlabel = "numbers from Google/Wikipedia",
       ylabel = "Number of deaths");

fig.savefig('../images/covid-plot.png')
```

```python
fig, ax = plt.subplots()
ax.barh(list(covid_deaths.keys()), list(covid_deaths.values()))
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots


```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                               ncols = 2,
                                               figsize = (10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.barh(list(covid_deaths.keys()), list(covid_deaths.values()));
ax4.hist(np.random.rand(1000));
```

```python
# Subplots option 1 
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].barh(list(covid_deaths.keys()), list(covid_deaths.values()));
ax[1, 1].hist(np.random.rand(1000));
```

## Plotting from Pandas DataFrames

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales = pd.read_csv("../data/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index = pd.date_range("11/05/2020", periods = 1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("12/05/2020", periods = len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x = "Sales Date", y = "Total Sales");
```

```python
# Reassign price column to int
car_sales["Price"] = car_sales["Price"].astype(int)

# Plot scatter plot with price column as numeric
car_sales.plot(x = "Odometer (KM)", y = ("Price"), kind = "scatter");
```
